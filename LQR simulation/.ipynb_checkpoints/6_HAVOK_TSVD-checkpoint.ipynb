## Low-rank approximation using mixed-integer convex optimisation framework

```python
U, S, Vt = svd(model1.H, full_matrices=False)
m, n = model1.H.shape[0], model1.H.shape[1]
M = min(m,n)
alpha = 1 # Relative importance of nb. of SVs

mu = cp.Variable(M,integer=True) # Sparsity promoting variable
Sk = cp.diag(S)@cp.diag(mu)
H = U@np.diag(S)@Vt # Original matrix (reduced)
Hk = U@Sk@Vt # Low rank approximation
print(S.shape, mu.shape, M)

objective = cp.Minimize(cp.norm(H - Hk,'fro') + alpha*cp.trace(Sk))

constraints = []
gamma = cp.multiply(mu,S)
print(gamma.shape)
gamma_prime = gamma
gamma_prime[:M-2] = gamma[1:M-1]
gamma_prime[M] = 0
constraints.append(cp.multiply(mu,S) >= gamma_prime)

prob = cp.Problem(objective, constraints)
prob.solve(solver=cp.MOSEK, verbose=True)
```
